**Importing necessary libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
%matplotlib inline
```

**Importing data & displaying it**

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
stud_data = pd.read_csv(url)

stud_data
```

**Plotting Score Distribution**

```python
stud_data.plot(x='Hours', y='Scores', style='*')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

**Indexing the data**

```python
X = stud_data.iloc[:, :-1].values  
y = stud_data.iloc[:, 1].values
```

**Splitting the data into train and test**

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.25,random_state=0) 
```

**Making an instance of the model and fitting the values for x & y**

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")
```

**Plotting data points along with regression line**

```python
rline = regressor.coef_*X+regressor.intercept_
plt.scatter(X, y,15)
plt.plot(X, rline);
plt.show()
```

**Predicting values using the test set**

```python
print(X_test) # Test set
y_pred = regressor.predict(X_test) # Predicting the scores
```

**Comparing Actual and Predicted values from the dataset**

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

**What will be predicted score if a student studies for 9.25 hrs/ day?**

```python
hours = 9.25 #given 
own_pred = regressor.predict(np.array([hours]).reshape(-1,1))
print("No. of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

**Calculating Mean absolute error for our model**

```python
from sklearn import metrics  
print('Mean Absolute Error is:', 
      metrics.mean_absolute_error(y_test, y_pred))
```
